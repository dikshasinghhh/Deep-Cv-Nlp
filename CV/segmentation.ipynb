```python
import cv2
import os
import matplotlib.pyplot as plt

# Path to folder
folder = '/kaggle/input/iris-computer-vision/iris-virginica'

# List all image files
image_files = [f for f in os.listdir(folder) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
print(f"Found {len(image_files)} images.")

# Process first 5 images for demonstration
for img_file in image_files[:5]:
    img_path = os.path.join(folder, img_file)
    
    # Read grayscale image
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"⚠️ Could not read {img_file}")
        continue
    
    # Resize image
    img = cv2.resize(img, (256, 256))
    
    # Apply Gaussian blur for noise reduction
    blur = cv2.GaussianBlur(img, (5, 5), 0)
    
    # Otsu's thresholding
    ret, thresh = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    
    # Display
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title('Original')
    plt.axis('off')
    
    plt.subplot(1, 2, 2)
    plt.imshow(thresh, cmap='gray')
    plt.title("Otsu Segmentation")
    plt.axis('off')
    
    plt.show()
```
