```python
import cv2
import os
import matplotlib.pyplot as plt

# Path to your dataset folder
dataset_path = '/kaggle/input/iris-computer-vision/iris-setosa'

# Get all image file names
image_files = [f for f in os.listdir(dataset_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Check how many images found
print(f"Found {len(image_files)} images in dataset.")

# Create a folder to save the edge-detected images (optional)
output_folder = '/kaggle/working/iris-edges'
os.makedirs(output_folder, exist_ok=True)

# Loop through images and apply edge detection
for filename in image_files:
    img_path = os.path.join(dataset_path, filename)
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # convert to grayscale
    
    # Apply Canny edge detection
    edges = cv2.Canny(img, 100, 200)
    
    # Save edge image
    edge_path = os.path.join(output_folder, f"edge_{filename}")
    cv2.imwrite(edge_path, edges)

    # Display one example
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.title('Original')
    plt.imshow(img, cmap='gray')
    plt.axis('off')

    plt.subplot(1,2,2)
    plt.title('Edges')
    plt.imshow(edges, cmap='gray')
    plt.axis('off')

    plt.show()
    break  # remove this line if you want to display all images
```
